```python
from scipy import stats 
DataFrame= pd.crosstab(df['Gender'], df['Specialization'])
# Display the contingency table
print("DataFrame:")
print(DataFrame)
# Perform Chi-Square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(DataFrame)
# Create a results DataFrame with reset index
results = pd.DataFrame({
'Metric': ['Chi-Squared Statistic', 'P-value', 'Degrees of Freedom','Conclusion'],
'Value': [chi2_stat,p_value,dof,"Reject the null hypothesis" if p_value < 0.05 else "Fail to reject theâ£null hypothesis"]
})
# Reset the index of the results DataFrame
results.reset_index(drop=True, inplace=True)
print("\nChi-Square Test Results:")
print(results)
```
